```python
import os
from pydub import AudioSegment

# Directory containing the .wav files
directory = './stimuli/syllables_raw'
output_directory = './output'


# Target duration in milliseconds
target_duration_ms = 350
# Convert target duration to number of samples

if not os.path.exists(output_directory):
    os.makedirs(output_directory)
    print(f"Directory '{output_directory}' created.")
else:
    print(f"Directory '{output_directory}' already exists.")

# Function to trim the first part of the audio
def trim_wav(file_path, target_duration):
    # Load the audio file
    audio = AudioSegment.from_wav(file_path)
    
    # Calculate the number of samples to trim
    duration_samples = len(audio)
    print(duration_samples)
    # If the audio is longer than the target duration, trim it
    if duration_samples > target_duration:
        trimmed_audio = audio[(duration_samples-target_duration):]  # Keep only the first target_duration_ms milliseconds
        print(len(trimmed_audio))
    else:
        trimmed_audio = audio  # If it's already shorter, keep the whole file
    
    return trimmed_audio

# Iterate through all files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".wav"):
        file_path = os.path.join(directory, filename)
        
        # Trim the file to 320ms
        trimmed_audio = trim_wav(file_path, target_duration_ms)
        
        # Save the trimmed audio to the output directory
        output_file_path = os.path.join(output_directory, filename)
        trimmed_audio.export(output_file_path, format="wav")
        
        print(f"Trimmed {filename} to {target_duration_ms}ms and saved to {output_file_path}")
```
